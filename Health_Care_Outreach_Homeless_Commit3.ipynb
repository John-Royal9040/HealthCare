Revised Research Question

How does education, income levels, and awareness impact the utilization of Medicaid and other healthcare benefits among individuals in urban areas, specifically homeless population in Baltimore,Maryland, who have access to these resources? Additionally, what factors contribute to the underutilization of available healthcare programs despite their accessibility?

Purpose Statement

The purpose of this study is to analyze the impact of education, income levels, and awareness on the utilization of Medicaid and other healthcare benefits among individuals residing in urban areas with access to these resources. By examining patterns of benefit usage, this research seeks to identify key factors that influence enrollment and participation in available healthcare programs. Additionally, this study aims to explore the reasons behind the underutilization of these programs despite their accessibility with comparissons to homeless populations within Baltimore, Maryland. Understanding these dynamics will provide valuable insights for policymakers, healthcare providers, and community organizations to improve outreach, education, and program effectiveness, ultimately enhancing healthcare equity and accessibility.
